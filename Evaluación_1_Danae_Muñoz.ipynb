**Clasificación de Variables**

* **Categórico Nominal:** Categorías sin un orden o jerarquía específica.
  * Ejemplo: Country, Region, WineStyle, Winery.

* **Categórico Ordinal:** Categorías que siguen un orden lógico o escala.
  * Ejemplo: (En este dataset no se identifican variables ordinales explícitas)

* **Numérico Discreto:** Valores que surgen de un conteo y no admiten decimales.
  * Ejemplo: NumberOfRatings, Year.

* **Numérico Continuo:** Valores que resultan de una medición y pueden tener decimales infinitos.
  * Ejemplo: Price, Rating.

```python
pip install category_encoders
```

```python
from sklearn.preprocessing import OneHotEncoder
import category_encoders as ce
```

```python
# Aplicar OneHotEncoder a Country y WineStyle
# Esto convierte esas columnas categóricas en columnas numéricas.
ohe = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

ohe_encoded = ohe.fit_transform(df_wines[['Country', 'WineStyle']])
```

Esto convierte las columnas categóricas Country y WineStyle en columnas numéricas.

```python
# Convertir este resultado a DataFrame
ohe_df = pd.DataFrame(
    ohe_encoded,
    columns=ohe.get_feature_names_out(['Country', 'WineStyle']),
    index=df_wines.index
)
```

Esto transforma el resultado del encoder en una tabla normal con nombres de columnas claros.

```python
# Aplicar Binary Encoding a Region
be = ce.BinaryEncoder(cols=['Region'])

region_encoded = be.fit_transform(df_wines[['Region']])
```

Convierte Region en varias columnas numéricas, pero sin generar 861 columnas como haría OneHot.

```python
# Unir todo en un nuevo dataframe final
df_final = pd.concat(
    [
        df_wines.drop(columns=['Country', 'WineStyle', 'Region']),
        ohe_df,
        region_encoded
    ],
    axis=1
)
```

Para poder utilizar las variables categóricas en el modelo, fue necesario transformarlas a formato numérico mediante técnicas de encoding.

En el caso de las variables Country y WineStyle, se aplicó One-Hot Encoding, ya que cuentan con un número reducido de categorías. Este método consiste en crear una columna por cada categoría, permitiendo representar cada valor de forma independiente sin generar conflictos en el modelo.

Por otro lado, la variable Region presenta una gran cantidad de categorías (alta cardinalidad). Si se aplicara One-Hot Encoding, se generarían demasiadas columnas, lo que afectaría el rendimiento del modelo. Por esta razón, se utilizó Binary Encoding, una técnica que permite representar las categorías utilizando un menor número de columnas mediante combinaciones binarias, logrando así reducir la dimensionalidad del dataset.
